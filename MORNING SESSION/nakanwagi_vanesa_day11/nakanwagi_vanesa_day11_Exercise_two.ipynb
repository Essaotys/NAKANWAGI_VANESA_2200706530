```python
#describing the data
import pandas as pd
data=pd.read_csv("heart.csv")
print(data.describe())
```

```python
#Checking for null values
print(data.isnull().sum())
```

```python
#Checking for duplicates 
data=pd.DataFrame(data)
duplicates = data.duplicated(keep=False)

# print the duplicate rows
print(data[duplicates])
```

```python
#Removing duplicates
remove_duplicates = data.drop_duplicates()
print(remove_duplicates)
```

```python
#Exploratory data analysis
#mean ,mode,and median
print("Mean:\n",data.mean())
print()
print("Mode:\n",data.mode().iloc[0])

print("Median:\n",data.median())
print()


```

```python
#Range.
print("Range:\n", data.max() - data.min())
```

```python
#Variance
print("Variance:\n", data.var())
```

```python
#Standard deviation 
print("Standard Deviation:\n", data.std())
```

```python
#Iterquartile mean
print("Iterquartile Range:\n", data.quantile(0.75) - data.quantile(0.25))
```

```python
#Visualization
#Histogram
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for Age
plt.figure(figsize=(5, 5))
sns.histplot(data['age'], kde=True, bins=10)
plt.title('Distribution Of Age')
plt.xlabel('age')
plt.ylabel('sex')
plt.show()
```

```python
#Heatmap
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 10))
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
#Bar graph
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 4))
sns.countplot(x='sex', data=data, palette='viridis')
plt.title('Distribution of Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.show()
```

```python

# Scatter plot examining the relationship between age and cholesterol
plt.figure(figsize=(5, 5))
sns.scatterplot(x='age', y='chol', data=data)
plt.title('Scatter Plot of Age versus Cholesterol')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.show()
```
